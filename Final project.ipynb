```python
from keras.models import model_from_json
import cv2
import numpy as np


# Step 1: Load the Model from Json File
json_file = open('./model.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)

# Step 2: Load the weights
loaded_model.load_weights("./model.h5")
print("Loaded model from disk")

# Step 3: Compile the model
loaded_model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
total=0
correct=0
images = []
for img_path in glob.glob('D:/new/male/*.jpg'):
    images.append(mpimg.imread(img_path))

plt.figure(figsize=(20,10))
columns = 5
for i, image in enumerate(images):
    total+=1
    plt.subplot(len(images) / columns + 1, columns, i + 1)
    plt.axis("off")
    plt.imshow(image) 
   # np.resize(img, (-1, <image shape>)
    image= cv2.resize(image,(50,50))
    image = np.expand_dims(image, axis=0)#to make array as 4 dimension since convolution is of 4d
    
    result = loaded_model.predict_classes(image)
    if(result[0][0] == 1):
        correct+=1
        print("We guess this must be a Male!")
    else:
        #correct+=1
        print("We guess this must be a Female!")
    
    #plt.imshow(image)
print("Total number of male images in testing dataset")
print(total)
print("Correctly predicted number of male images in testing dataset")
print(correct)
accuracy=correct*100/total
print("Accuracy of male test images")
print(accuracy)
```
